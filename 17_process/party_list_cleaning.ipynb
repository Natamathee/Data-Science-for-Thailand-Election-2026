```python
import pandas as pd

df = pd.read_csv("../output/raw_csv/17_party_list.csv")
print(df.shape)
print(df.dtypes)
print()
df.head(20)
```

```python

party_ref = pd.read_csv("../reference/party_list_ref.csv", dtype={"หมายเลข": int})

# Map หมายเลข → ชื่อพรรค
party_map = dict(zip(party_ref["หมายเลข"], party_ref["พรรค"]))
df["พรรค"] = df["หมายเลข"].map(party_map)

print(df[["หมายเลข", "พรรค", "คะแนน"]].head(20))
print("null พรรค:", df["พรรค"].isna().sum())
```

```python
df.head()
```

```python
df.to_csv("../output/cleaned/17_party_list.csv", index=False, encoding="utf-8-sig")
print(f"✅ Export {len(df)} rows → output/cleaned/17_party_list.csv")
```
